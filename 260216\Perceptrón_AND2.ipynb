```python
# importar librería numpy
import numpy as np
```

```python
class Perceptron(object):
  def __init__(self, learning_rate=0.01, n_iterations=100):
    self.learning_rate = learning_rate
    self.n_iterations = n_iterations
    self.weights = None
    self.bias = None

  def fit(self, X, y):
    n_samples, n_features = X.shape

    #inicializar pesos y sesgo (bias) con ceros
    self.weights = np.random.rand(n_features)
    self.bias = 0

    # Etiquetas y deben ser 1 o -1
    y_ = np.array([1 if i > 0 else -1 for i in y]) # Corrected 'fpr' to 'for' and '1 > 0' to 'i > 0'

    # Proceso de aprendizaje
    for _ in range(self.n_iterations):
      for idx, x_i in enumerate(X):
        condition = y_[idx] * (np.dot(x_i, self.weights) + self.bias) >= 0
        if not condition:
          self.weights += self.learning_rate * y_[idx] * x_i
          self.bias += self.learning_rate * y[idx]

  def predict(self, X): # Corrected indentation
    linear_output = np.dot(X, self.weights) + self.bias
    predicted_labels = np.where(linear_output >= 0, 1, -1)
    return predicted_labels
```

```python
# Ejemplo del uso del perceptrón
import numpy as np

# Datos de entrenamiento (X) y etiquetas (y)
X = np.array([
    [-1, -1],
    [-1,  1],
    [ 1, -1],
    [ 1,  1]
])

y = np.array([-1, -1, -1, 1]) # Modificado para representar la compuerta AND
```

```python
# Crear y entrenar el perceptrón
perceptron = Perceptron(learning_rate=0.01, n_iterations=100)
perceptron.fit(X, y)
```

```python
# Predecir las etiquetas de clase
predictions = perceptron.predict(X)
print("Predicciones: ",predictions)
```
